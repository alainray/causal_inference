```
import matplotlib.pyplot as plt 
import random

n_samples = 100
x = [random.random()*2 + 1 for x in range(n_samples)]
y = [2*x_i + random.random() - 0.5 for x_i in x]
# line 1 points 
x1 = [1,2,3] 
y1 = [2,4,6] 

x2 = [1,2,3] 
y2 = [1.5,3.5,5.5] 


x3 = [1,2,3] 
y3 = [2.5,4.5,6.5] 
```

```
def plot_curves(x,y, labels, scatter_x, scatter_y, reverse=False):
  for x_i,y_i,l in zip(x,y,labels):
      if not reverse:
        plt.plot(x_i, y_i, label=l)
      else:
        plt.plot(y_i,x_i,label=l)
  if not reverse:
      plt.scatter(scatter_x, scatter_y)
      plt.title('X-->Y') 
  else:
      plt.scatter(scatter_y, scatter_x)
      plt.title('Y-->X') 
  # naming the x axis 
  plt.xlabel('x - axis') 
  # naming the y axis 
  plt.ylabel('y - axis') 
  # giving a title to my graph 
  
  # show a legend on the plot 
  plt.legend() 
  # function to show the plot 
  plt.show()
```

```
plot_curves([x1,x2,x3], [y1,y2,y3], ['avg','lower_bound','upper_bound'], x, y)
```

```
plot_curves([x1,x2,x3], [y1,y2,y3], ['avg','upper_bound','lower_bound'], x, y, reverse=True)
```

```
import numpy as np
n_samples = 100
x = [random.random()*2 + 1 for x in range(n_samples)]
y = [x_i*x_i + random.random() - 0.5 for x_i in x]
# line 1 points 
delta = 0.01
x1 = np.arange(1,3 + delta, delta) 
y1 = [x_i*x_i for x_i in x1]
x2 = x1
y2 = [x_i*x_i -0.5 for x_i in x1]
x3 = x1 
y3 = [x_i*x_i + 0.5 for x_i in x1]
```

```
plot_curves([x1,x2,x3], [y1,y2,y3], ['avg','lower_bound','upper_bound'], x, y)
```

```
plot_curves([x1,x2,x3], [y1,y2,y3], ['avg','upper_bound','lower_bound'], x, y, reverse=True)
```
